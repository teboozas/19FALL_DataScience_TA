## 2.5 Module, Package, and `import` command

### Module and Package

Now we just finished all the basics in OOP via Python language, dividing into type, attribute, and method.

Programmers usually write codes in this way with Python, and store them into **single script file**. And scripts with similar usage are gathered into **single folder**. These are called **module** and **package**, respectively.

**Module = `.py` format script(file)**

Module is a Python script stored in `.py` format. Programmers usually store classes, functions, variables with similar functionality. Module and included objects are referable in other Python script or Jupyter Notebook environment.


**Package = folder which `.py` files are stored**

Packages are kind of folder tree in Python. Usually, similar modules are stored and compose a single package. The way to turn a certain folder into a Python package is making empty module named `__init__.py`. This is why you often can find `__init__.py` file in open-source storage like GitHub.

Subpackages are subfolder stored in a certain Python package. These are also defined with `__init__.py` module stored into.

`__init__.py` module can be used to initialization of the package at the beginning of running. Objects defined in `__init__.py` are refered with import of the package.

**Dot syntax to refer package and module**

Dot syntax, used to inclusion relationship of object-attribute/object-method, can be exploited to refer objects in package and module. For example:

> `package.subpackage.class.method`

Several names linked with dot represent inclusion of objects. This is why dot syntax must be handled carefully.

### `import` command - importing package and module

Like other languages, `import` command is used to import package and module in Python

**`import ... as ...`**

`import` command refers package, stored in library folder suited for a certain Python version. You can use `as` command to import packages into another name.

There are traditions in name changing for famous Python packages like:

> `import numpy as np`<br>
> `import pandas as pd`<br>
> `import tensorflow as tf`

```python
# most commonly used 'import' command in Python programming for data science
import numpy as np
```

**`from ... import ...`**

If you want to import specific object or methods from package/module, `from` command is need. `from` command specifies package or module that target object is stored.

Special built-in package named by `__future__` contains functionalities that are not officially supported in current version of Python. You can use these functionalities `__future__` package like:

> `from __future__ import print_function`

```python
from __future__ import print_function
```
