Most frequently and widely used packages like Numpy, Scipy, Pandas, Scikit-learn, Matplotlib are all members of SciPy ecosystem, and constantly expanding their area with active feedbacks and updates.

## 3.2 NumPy

`NumPy` was introduced for numerical computing in Python. **NumPy array**(`ndarray`) data type is the core of NumPy package, which enables fast, stable, and effective calculation for massive data with high-dimensionality. That's why NumPy is the essential package in data science.

* `ndarray` data type supports standard methods for handling sequence object like `range()`, list comprehension.
* NumPy package also privides various linear algebra functions to handle matrix or tensor-coded data(dot product, eigenvalue decomposition, etc.)

```python
import numpy as np
dir(np)
```

## 3.3 Pandas

`Pandas` is a package to support *labeled Column-oriented data* like other data handling frameworks.

* `DataFrame` is a data type provided in `Pandas`, which has the same functionality and characteristic with data frame in R-language.
* `DataFrame` also contains useful built-in methods. So various works like preprocessing, computing simple statistic can easily done with them.

```python
import pandas as pd
dir(pd)
```

## 3.4 Matplotlib

`Matplotlib` provides MATLAB-style graphics, to visualize data and their feature and structure.

* `Matplotlib` is compatible with data types in `NumPy` or `Pandas`.
* Also `Matplotlib` supports `ggplot` style graphics like R-language.

```python
import matplotlib.pyplot as plt
dir(plt)
```

## 3.5 Scikit-learn

In `Scikit-learn` package, most of the widely-used machine learning techniques are implemented, including major classification and regression methods.

* To work with `Scikit-learn`, users usually import subpackages or modules matched with purpose, rather than import entire package itself.

* In `Scikit-learn` package, major supervised learning, unsupervised learning, model evaluation, data preprocessing methods are included, and regularly updated by numerous contributors in the world.

    * **Supervised learning** - linear regression, support vector machine, naive Bayes classifier, k-nearest neighbor(KNN), etc.
    * **Unsupervised learning** - principle component analysis(PCA), k-means clustering, etc.
    * **Model evaluation** - MSE, MAE, V-measure, cross-validation, etc.

```python
from sklearn import neighbors
knn = neighbors.KNeighborsClassifier
```
