## 4.1 Clone GitHub repository

Many Python programmers prefer to store their codes and data in GitHub repository(folder)

To access them, you can use `clone` shell command provided in Jupyter Notebook(and Colab). This command copies target repository and moves it on the personal storage. Whit cloning repository, you can use all of the codes and data like your own.

## 4.2 Open `.ipynb` notebook stored in Github repository

Colab and GitHub can be directly linked to open or store notebooks in `.ipynb` format. You can open notebooks in GitHub repository by URL, and execute them in 'playground mode'. This is a powerful advantage of Colab

To open notebook with URL, follow a process below:

(sample URL - https://github.com/jakevdp/PythonDataScienceHandbook)

> **'File' → 'Open notebook' → 'GITHUB' tab → enter GitHub repository URL → press 'Enter' → open `.ipynb` file**

## 4.3 Store `.ipynb` notebook in Github repository

If you access to GitHub with your account, you can directly *commit*(store) `.ipynb` notebook file that is written in Colab.

> **'File' → 'Save a copy in GitHub' → check related settings and press 'OK'**
